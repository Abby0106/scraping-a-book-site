```python
from IPython.display import display
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup

r = requests.get("https://books.toscrape.com/")
soup = BeautifulSoup(r.content,'lxml')


print(r.status_code)

book_details = soup.find_all('article', attrs={'class':'product_pod'})
print(len(book_details))

def get_book_detail(book_details):
    title = book_details.find('h3').find('a')['title']
    price = book_details.find('p', attrs={'class':'price_color'}).text
    rating = book_details.find('p', attrs={'class':'star-rating'})['class'][-1]
    return title,price,rating
print(get_book_detail(book_details[0]))
b = get_book_detail(book_details[1])

def fix_price(price):
    c_price = float(re.sub('[^0-9.]',"",price))
    return c_price
print(fix_price(b[1]))

def fix_rating(rating):
    map_rating = {'One':1,
                  'Two':2,
                  'Three':3,
                  'Four':4,
                  'Five':5,
                  'Six':6}
    return map_rating[rating]
print(fix_rating(b[-1]))

def get_book_detail(book_details):
    title = book_details.find('h3').find('a')['title']
    price = book_details.find('p', attrs={'class':'price_color'}).text
    rating = book_details.find('p', attrs={'class':'star-rating'})['class'][-1]
    return {'title':title,
            'price':fix_price(price),
            'rating':fix_rating(rating)}
bd = get_book_detail(book_details[0])
print(bd)

All_detail = [get_book_detail(book_detail) for book_detail in book_details]
# for detail in All_detail:
#     print(detail)

df = pd.DataFrame(All_detail)

display(df)





























```
